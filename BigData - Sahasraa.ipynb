```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Read the CSV directly
df = pd.read_csv("training_data.csv")

# Create a label encoder
le = LabelEncoder()

# Loop through columns and transform if they are of object type or have mixed types
for col in df.columns:
    if df[col].dtype == 'object':
        # Convert entire column to string type to handle mixed data types
        df[col] = df[col].astype(str)
        df[col] = le.fit_transform(df[col])
```

```python
all_the_columns = df.columns.drop(['Claim_Count', 'Loss_Amount', 'Frequency', 'Severity', 'Loss_Ratio'])
```

```python
all_the_columns
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import math
import os

# Load the training data

# 1. Creating Training Portfolios:
def create_portfolios(data, num_policies, loss_percentage):
    # Filter policies with losses
    loss_data = data[data["Loss_Ratio"] > 0].sample(frac=loss_percentage)
    no_loss_data = data[data["Loss_Ratio"] == 0].sample(n=num_policies-len(loss_data))
    
    # Create portfolio
    portfolio = pd.concat([loss_data, no_loss_data])
    
    return portfolio

def calculate_ln_LR(portfolio):
    total_premium = portfolio["Annual_Premium"].sum()  # updated column name
    total_losses = portfolio["Loss_Amount"].sum()

    return math.log(total_losses / total_premium)

# 2. Feature Engineering:
def derive_features(portfolio):
    # Using Driver_Minimum_Age and Driver_Maximum_Age to compute an average age
    mean_age = (portfolio["Driver_Minimum_Age"] + portfolio["Driver_Maximum_Age"]) / 2.0
    mean = []
    for i in all_the_columns:
        mean.append(portfolio[i].mean())
    return mean,mean_age.mean()  # Added mean() for mean_age as it's a Series now

# 3. Model Training:
features = []
targets = []

for i in range(100):  # e.g., create 100 training portfolios
    portfolio = create_portfolios(df, 1000, 0.05)  # for simplicity
    mean,mean_age = derive_features(portfolio)
    ln_LR = calculate_ln_LR(portfolio)
    
    features.append(mean+[mean_age])
    targets.append(ln_LR)

X_train, X_val, y_train, y_val = train_test_split(features, targets, test_size=0.2)

model = RandomForestRegressor()
model.fit(X_train, y_train)

val_preds = model.predict(X_val)
mse = mean_squared_error(y_val, val_preds)
print(f"Validation MSE: {mse}")
```

```python
# 4. Testing:

def preprocess(df):
    # Create a label encoder
    le = LabelEncoder()

    # Loop through columns and transform if they are of object type or have mixed types
    for col in df.columns:
        if df[col].dtype == 'object':
            # Convert entire column to string type to handle mixed data types
            df[col] = df[col].astype(str)
            df[col] = le.fit_transform(df[col])
    return df

test_features = []

testing_path = "testing_portfolios"
files = [f for f in os.listdir(testing_path) if f.endswith('.csv')]

for file in files:
    file_path = os.path.join(testing_path, file)
    test_portfolio = pd.read_csv(file_path)
    test_portfolio = preprocess(test_portfolio)
    mean,mean_age = derive_features(test_portfolio)
    test_features.append(mean+[mean_age])

test_preds = model.predict(test_features)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import math
import os

# Load the training data
train_data = pd.read_csv("training_data.csv")

def create_portfolios(data, num_policies, loss_percentage):
    # Filter policies with losses
    loss_data = data[data["Loss_Ratio"] > 0].sample(frac=loss_percentage)
    no_loss_data = data[data["Loss_Ratio"] == 0].sample(n=num_policies-len(loss_data))
    
    # Create portfolio
    portfolio = pd.concat([loss_data, no_loss_data])
    
    return portfolio

def derive_features(portfolio):
    # Updated to use the correct columns from the dataset
    mean_age = portfolio["Driver_Maximum_Age"].mean()
    mean_miles = portfolio["Vehicle_Miles_To_Work"].mean()
    
    return mean_age, mean_miles

def calculate_ln_LR(portfolio):
    total_premium = portfolio["Annual_Premium"].sum()  # updated column name
    total_losses = portfolio["Loss_Amount"].sum()
    return math.log(total_losses / total_premium)

# Training:
features = []
targets = []

for i in range(100):  # e.g., create 100 training portfolios
    portfolio = create_portfolios(train_data, 1000, 0.05)  # for simplicity
    mean_age, mean_miles = derive_features(portfolio)
    ln_LR = calculate_ln_LR(portfolio)
    
    features.append([mean_age, mean_miles])
    targets.append(ln_LR)

X_train, X_val, y_train, y_val = train_test_split(features, targets, test_size=0.2)

model = RandomForestRegressor()
model.fit(X_train, y_train)

val_preds = model.predict(X_val)
mse = mean_squared_error(y_val, val_preds)
print(f"Validation MSE: {mse}")

# Testing:
test_features = []

testing_path = "testing_portfolios"
files = [f for f in os.listdir(testing_path) if f.endswith('.csv')]

for file in files:
    file_path = os.path.join(testing_path, file)
    test_portfolio = pd.read_csv(file_path)
    mean_age, mean_miles = derive_features(test_portfolio)
    test_features.append([mean_age, mean_miles])

test_preds = model.predict(test_features)
print("Test Predictions:", test_preds)
```
